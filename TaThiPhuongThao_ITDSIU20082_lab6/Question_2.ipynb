```python
from sklearn.multiclass import OneVsRestClassifier
# Load dataset
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data
target = iris.target
```

```python
from sklearn.model_selection import train_test_split
datasets = train_test_split(features, target, test_size=0.2)

X_train, X_test, y_train, y_test = datasets
```

```python
from sklearn.svm import SVC
import numpy as np

#SVC Model
model = SVC(gamma='scale',random_state=0)

#Define 1-vs-Rest Strategy / Classifier
ovr = OneVsRestClassifier(model)

#fit model to training data
ovr.fit(X_train, y_train)

#Predications
ovr_pred = ovr.predict(X_test)
```

```python
#Adding Predictions to Test Dataset
import pandas as pd
ovr_df = X_test.copy()
df = pd.DataFrame(ovr_df, columns=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'])
```

```python
df['Actual'] = y_test
df['Predicted'] = ovr_pred
```
